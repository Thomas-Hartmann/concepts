### Read content from web page using the library: ["requests"](http://docs.python-requests.org/en/latest/user/install/#install) AND [beatufulsoup4](https://www.crummy.com/software/BeautifulSoup/)

```python
# Challenge 3
import requests
from bs4 import BeautifulSoup as soup
from bs4 import Comment
response = requests.get("http://www.pythonchallenge.com/pc/def/ocr.html")
t = soup(response.text)
print(t) # available: r.text, r.json(, r.raw, r.content (bytes)
print("----------------------------------------------------------------")
# print all the <li> elements
test_li = t.find_all('li')
print(test_li)
# get all first comment:
comment = t.find_all(String=lambda text:isinstance(text, Comment))
print(comment)
```

```python
import requests
from bs4 import BeautifulSoup as soup
from bs4 import Comment
from collections import defaultdict
response = requests.get("http://www.pythonchallenge.com/pc/def/ocr.html")
t = response.text
#print(t)
s_t = soup(response.text, "html.parser")
comment = s_t.find_all(string=lambda text: isinstance(text, Comment))[1]
print(comment)
dict = defaultdict(int)
for char in comment:
    dict[char] += 1
#print(dict)
sorted_list = sorted(dict.items(), key=lambda pair: pair[1]) # items is a list of key,value pairs sorted  by value (sorting by key would be lambda pair: pair[0])
print(sorted_list[:8])
result = [item[0] for item in sorted_list[:8]] # create a list of keys from the list of key,value pairs using a list comprehension
print("".join(result))

```
